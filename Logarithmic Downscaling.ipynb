# Different Downscaling Methods in Machine Learning

## 1) Logarithmic Downscaling

```python
import math
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import ticker
```

```python
x = [8056, 94561,32187,12687891,2,664,1237,1,3,4578,218987,12387891231,74484123,7894235488945321]
y = [10,122,385,144,205,456,500,1,39,100,191,262,413,314]
```

```python
logarithm = []
print("By default log to the base 'e' is considered and the value of 'e' -> 2.718281828…\n")
print("Original Values ","|     Log Reduced Values")
print("")

for i in x:
    g = math.log(i)
    print("  ", i, "         |      ", g)
    logarithm.append(g)
```

```python
sns.barplot(x = y, y = x)
plt.title('     Graph of Original Values')
plt.xlabel('Original Values')
plt.show()
print("The above Graph itself shows us that, the value at 314 is soo large in magnitude that the other values aren't even comparable and that's exacty what we're solving using Logarithm.")

print("\n-----------------------------------------------------------------------------------------------------\n")

print("THe below Graph shows, us how the Values have been downscaled, so that all the values are visually comparable.")
sns.barplot(y = logarithm, x = y)
plt.title('Graph of Logarithm Values')
plt.ylabel('Logarithm Values')
plt.show()
plt.show()
```

## 2) Standarization

```python
n = len(x)
mean_ans = 0
ans = 0
j = 0

for i in x:
    j = j + i
    k = i*i;
    ans = ans + k   
    
print("n : ", n)
print("Summation (X)   : ", j)
print("Summation (X^2) : ", ans)

mean = j/n

part_1 = ans/n
part_2 = mean*mean
standard_deviation = part_1 - part_2

print ("\nStandard Deviation : {:.5f}".format(standard_deviation))

print("Mean               : ", mean)

print("\n\n")

final_z_score = []

for i in x:
    z_score = (i-mean)/standard_deviation
    final_z_score.append("{:.32f}".format(z_score))


print("Y  |   Original DataSet   |               Z-Score ")
print()
for i in range (len(x)):
    print(y[i], " |  ", x[i], "              |     ", final_z_score[i])
    
    
print("\n\n\n VISUALIZATIONS")

fig = plt.figure()
fig.set_figwidth(8)
fig.set_figheight(6)
ax = fig.add_axes([0,0,1,1])

plt.title('Graph of Original Values')
plt.xlabel('Original Y Values')
plt.ylabel('Original X Values')
ax.bar(y, x, 15)
plt.xticks(y)
plt.show()

print("The above graph is of the Out of Scale Visually Incomparable Values.")

print("\n----------------------------------------------------------------------------------------------------\n")

print("The below Graph is the Standardized Scaled Down Values which are comparable.")
print("Just like we read, understood and illustrated, the Standardized Values are those which downscales all the Features to a scale common to all the Feature Values.")

fig = plt.figure()
fig.set_figwidth(10)
fig.set_figheight(7)
ax = fig.add_axes([0,0,1,1])

plt.rc('axes', titlesize = 25) 
#plt.rc('axes', labelsize = 20)
#plt.rc('xtick', labelsize = 10) 
#plt.rc('ytick', labelsize = 15) 
plt.title('Graph of Standardized Values')
plt.xlabel('Original Y Values')
plt.ylabel('Standardized Values')
ax.bar(y, final_z_score, 15)
plt.xticks(y)
plt.show()
```

## 3) Normalization

```python
min_dataset_1 = min(x)
max_dataset_1 = max(x)

print("Min(dataset_1) : ", min_dataset_1 )
print("Max(dataset_1) : ", max_dataset_1 )



final_normalization = []

print("\nNormalization = (Current_Value - Min) / (Max - Min)\n")

for i in x:
    normalization = (i - min_dataset_1) / (max_dataset_1 - min_dataset_1 )
    final_normalization.append("{:.20f}".format(normalization))
    
    
print(" Y   |  Original DataSet   |         Normalization ")
print()
for i in range (len(x)):
    print(y[i], " |  ", x[i], "            |     ", final_normalization[i])
    
    
print("\n\n")
    
fig = plt.figure()
fig.set_figwidth(8)
fig.set_figheight(6)
ax = fig.add_axes([0,0,1,1])

plt.title('Graph of Original Values')
plt.xlabel('Original Y Values')
plt.ylabel('Original X Values')
ax.bar(y, x, 15)
plt.xticks(y)
plt.show()

print("The above graph is of the Out of Scale Visually Incomparable Values.")

print("\n----------------------------------------------------------------------------------------------------\n")

print("The below Graph is the Normalized Scaled Down Values which are comparable.")
print("Just like we read, understood and illustrated with example, the Normalized Values also are scaled down but here it's in a range from 0 - 1.")

fig = plt.figure()
fig.set_figwidth(10)
fig.set_figheight(7)
ax = fig.add_axes([0,0,1,1])

plt.rc('axes', titlesize = 25) 
#plt.rc('axes', labelsize = 20)
#plt.rc('xtick', labelsize = 10) 
#plt.rc('ytick', labelsize = 15) 
plt.title('Graph of Normalized Values')
plt.xlabel('Original Y Values')
plt.ylabel('Normalized Values')
ax.bar(y, final_normalization, 15)
plt.xticks(y)
plt.show()    
```
